```python
import tensorflow as tf
from keras.preprocessing.image import img_to_array, load_img
import matplotlib.pyplot as plt
from keras.models import load_model
from keras.config import enable_unsafe_deserialization
import keras

keras.config.enable_unsafe_deserialization()
model = load_model('models/mult.keras')  # Replace with your model path
```

```python
test_image = load_img('dataset/images/0a172b0e-136b7a63.jpg', target_size=(256, 256)) 
test_image = img_to_array(test_image)

test_image = test_image.astype('float32') 
test_image /= 255.0

test_image = test_image.reshape((1, 256, 256, 3))
```

```python
test_preds = model.predict(test_image)


test_preds = test_preds.reshape((256, 256, 1))

plt.imshow(test_preds[:, :, 0], cmap='gray')  # Assuming grayscale output
plt.colorbar()
plt.title('Inference Prediction')
plt.show()
```
